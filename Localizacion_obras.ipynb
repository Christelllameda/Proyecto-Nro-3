```python
def obtener_latitudes(direcciones):
    time.sleep(5)
    
    geolocator = Nominatim(user_agent="myGeocoder")
    latitudes = []

    for direccion in direcciones:
        location = geolocator.geocode(direccion)
        if location:
            latitudes.append(location.latitude)
        else:
            latitudes.append(None)  # Otra opción es usar None para direcciones no encontradas

    return latitudes
```

```python
def obtener_coordenadas(dataframe):
    
    time.sleep(50)
    
    geolocator = Nominatim(user_agent="myGeocoder")

    latitudes = []
    
    for index, row in dataframe.iterrows():
        if not pd.isna(row['Dirección']):
            # Si hay una dirección, intenta obtener coordenadas a partir de ella
            location = geolocator.geocode(row['Dirección'])
        elif not pd.isna(row['Municipio']):
            # Si no hay dirección, intenta obtener coordenadas a partir del municipio
            location = geolocator.geocode(row['Municipio'])
        else:
            # Si no hay dirección ni municipio, asigna NaN a las coordenadas
            location = None

        if location:
            latitudes.append(location.latitude)
        else:
            latitudes.append(None)
        

    dataframe['Latitud'] = latitudes
   
    return dataframe
```
